```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
metadata = pd.read_csv("Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv", quotechar='"')
metadata_filtered = metadata[metadata['Region'] != '']
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv", skiprows=4, quotechar='"')
df['2023'] = pd.to_numeric(df['2023'].replace({',': ''}, regex=True), errors='coerce')
plt.figure(figsize=(12,6))
sns.countplot(y='Region', data=metadata_filtered, palette='pastel', order=metadata_filtered['Region'].value_counts().index)
plt.title('Number of Countries by Region')
plt.xlabel('Number of Countries')
plt.ylabel('Region')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='IncomeGroup', data=metadata_filtered, palette='muted', order=metadata_filtered['IncomeGroup'].value_counts().index)
plt.title('Number of Countries by Income Group')
plt.xlabel('Income Group')
plt.ylabel('Number of Countries')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
country_codes = metadata[metadata['Region'] != '']['Country Code'].tolist()
df_countries = df[df['Country Code'].isin(country_codes)]
df_countries = df_countries[df_countries['2023'] > 0]
plt.figure(figsize=(12,6))
sns.histplot(df_countries['2023'], bins=30, color='skyblue', kde=True)
plt.title('Distribution of Country Populations in 2023')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.tight_layout()
plt.show()
```
